```python
from scipy import stats
import matplotlib.pyplot as plt

```

```python
 

# x is random points from a left-skewed normal distribution center 5 no negative values
x = stats.skewnorm.rvs(5, size=1000) + 5

plt.hist(x, bins=100)
```

```python
bc_x = stats.boxcox(x)
print(f'lambda {bc_x[1]}')

plt.hist(bc_x[0], bins=100)
```
