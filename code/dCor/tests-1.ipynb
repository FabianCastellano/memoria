```python
import numpy as np
import matplotlib.pyplot as plt


# example Henon map - create & plot data

x = np.linspace(0,1,1000)
y = np.sin(12*x)/2 + 0.5

plt.plot(x,y)
```

```python
y_noise = y + np.random.normal(0,0.1,1000)

plt.plot(x,y_noise)
```

```python
from scipy.spatial import distance

print(distance.correlation(x, y))
```

```python
from scipy.spatial import distance

print(distance.correlation(x, y_noise))
```
