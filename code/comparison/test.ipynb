```python
from img_comparison import compare_images
import matplotlib.pyplot as plt
import numpy as np

img_path = '../boxcox/img/1.png'
img = plt.imread(img_path)
bw_img = 0.299*img[:,:,0] + 0.587*img[:,:,1] + 0.114*img[:,:,2]
bw_img_noise = bw_img + 0.5 * np.random.random(bw_img.shape)

comp = compare_images(bw_img, bw_img_noise, 'Original', 'Noisy')
comp.show()
```
