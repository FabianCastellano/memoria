```python
import numpy as np
import matplotlib.pyplot as plt


# example Henon map - create & plot data

x = np.linspace(0,1,1000)
y = np.sin(12*x)/2 + 0.5

plt.plot(x,y)
```

```python
y_noise = y + np.random.normal(0,0.1,1000)

plt.plot(x,y_noise)
```

```python
from minepy import MINE

def print_stats(mine):
    print ("MIC", mine.mic())
    print ("MAS", mine.mas())
    print ("MEV", mine.mev())
    print ("MCN (eps=0)", mine.mcn(0))
    print ("MCN (eps=1-MIC)", mine.mcn_general())
    print ("GMIC", mine.gmic())
    print ("TIC", mine.tic())

mine = MINE(alpha=0.6, c=15, est="mic_e")
mine.compute_score(x, y)

print("Without noise:")
print_stats(mine)
```

```python
print( "With noise:")
mine.compute_score(x, y_noise)
print_stats(mine)
```
